```python
import numpy as np
import matplotlib.pyplot as plt
import math
from scipy.integrate import odeint

%matplotlib inline

plt.rcParams['figure.figsize'] = [10, 6]
plt.rcParams['axes.grid'] = True
```

```python
def gillespie(rates, events, beta, gamma, init, mu=None, delta=None, seed=None, T=200, imports=False):
    t = 0
    X, Y, Z = init
    N = X + Y + Z

    if delta:
        SIR_data = np.array([t, X, Y, Z, 0])
    else:
        SIR_data = np.array([t, X, Y, Z])
        
    np.random.seed(seed)

    def smallest_event(events):
        dts = np.array([])
        for event in events:
            if event == 0:
                dts = np.append(dts, math.inf)
            else:                
                
                dt = -1 / event * np.log(np.random.random_sample())
                dts = np.append(dts, dt)
        return(np.amin(dts), list(dts).index(np.amin(dts)))


    def run_event(event):
        x_op, y_op, z_op = event
        return X + x_op, Y + y_op, Z + z_op
    
    while t < T:
        if Y == 0 and not imports:
            break
        h_rates = []
        
        args = Args()
        
        args.beta = beta
        args.gamma = gamma
        args.mu = mu
        args.X = X
        args.Y = Y
        args.Z = Z
        args.N = N
        if delta:
            args.delta = delta
    
        for k in rates:
            h_rates.append(k(args))
        
        dt, index = smallest_event(h_rates)
        
        X, Y, Z = run_event(events[index])
        t += dt
        N = X + Y + Z
        
        if delta:                   
            if index == 6:
                SIR_data = np.vstack((SIR_data, [t, X, Y, Z, 1]))
            else:
                SIR_data = np.vstack((SIR_data, [t, X, Y, Z, None]))
        else:
            SIR_data = np.vstack((SIR_data, [t, X, Y, Z]))

    return SIR_data
```

```python
class Args: pass

def f1(args):
    return args.beta * args.X * args.Y / args.N

def f2(args):
    return args.gamma * args.Y

def f3(args):
    return args.mu * args.N

def f4(args):
    return args.mu * args.X

def f5(args):
    return args.mu * args.Y

def f6(args):
    return args.mu * args.Z

# imports
def f7(args):
    return args.delta * args.N

# meta population
def m1(args):
    return args.lambd * args.X

def m2(args):
    return args.gamma * args.Y

# lattice 
def l1(args):
    return args.beta * (args.X / args.N) * args.term

def l2(args):
    return args.gamma * args.Y
```

```python
def SIR_with_demography(y, t, beta, gamma, mu):
    S, I, R = y
    ds_dt = mu - beta * S * I - mu * S
    di_dt = beta * S * I - gamma * I - mu * I
    dr_dt = gamma * I - mu * R
    return ds_dt, di_dt, dr_dt
```

```python
rates_nodemo = [f1, f2]
rates_demo = [f1, f2, f3, f4, f5, f6]
rates_imports = [f1, f2, f3, f4, f5, f6, f7]

events_nodemo = [(-1, 1, 0), (0, -1, 1)]
events_demo = [(-1, 1, 0), (0, -1, 1), (1, 0, 0), (-1, 0, 0), (0, -1, 0), (0, 0, -1)]
events_imports = [(-1, 1, 0), (0, -1, 1), (1, 0, 0), (-1, 0, 0), (0, -1, 0), (0, 0, -1), (0, 1, 0)]
events_meta = [(-1, 1, 0), (0, -1, 1)]
```

### Negative Covariance

```python
covs = []
r0s = []

gamma = 1
mu = 1/60

threshold = 0

for beta in range(1, 16):
    t = 0
    localcovs = []
    while t < 100:
        A = gillespie(rates_demo, events_demo, beta, gamma, (299, 1, 0), mu, seed=18)
        cov_matrix = np.cov(np.stack((A[:,1], A[:,2]), axis=0), bias=True)
        covariance = cov_matrix[0][1]
        if(covariance < 0):
            localcovs.append(covariance)
        t+= 1
    covs.append(np.mean(localcovs))
    r0s.append(beta/gamma)
    print(beta, np.mean(localcovs))
```

```python
class Args: pass

def gillespie_meta(rates, events, beta, gamma, rho, init, maxT=3000, seed=3):
    # set sed
    np.random.seed(seed)
    
    # Find smallest event
    def smallest_event(events):
        dts = np.array([])
        for event in events:
            if event == 0:
                dts = np.append(dts, math.inf)
            else:                
                dt = -1 / event * np.log(np.random.random_sample())
                dts = np.append(dts, dt)
        return(np.amin(dts), list(dts).index(np.amin(dts)))
    
    # execute event    
    def run_event(event):
        x_op, y_op, z_op = event
        return x_op, y_op, z_op
    
    # Total amount of populations    
    meta_pops = len(init)
    
    t = 0
    
    # Initialize arrays    
    Ts = np.full((maxT), None)    
    X = np.full((meta_pops, maxT), None)
    Y = np.full((meta_pops, maxT), None)
    Z = np.full((meta_pops, maxT), None)    
    
    N = np.full((meta_pops), None)

    lambdas = np.empty((meta_pops, 1))
    
    # Get last added element in arrays    
    current = 0
    
    # init    
    for i in range(meta_pops):
        X[i][0], Y[i][0], Z[i][0] = init[i]
        N[i] = sum(init[i])
        
    while t < maxT and current < maxT - 1:   
        # check wheter there are still infected people in metapops
        e = 0
        for i in range(meta_pops):
            if(Y[i][current] == 0):
                e += 1
        if(e == meta_pops):
            break

        for i in range(meta_pops):
            lambda_n = 0
            for j in range(meta_pops):
                lambda_n += rho[i][j] * (Y[j][current] / N[i])
            lambdas[i] = beta[i] * lambda_n

        h_rates = []            
        args = Args()
        
        for i in range(meta_pops):
            args.lambd = float(lambdas[i])
            args.X = X[i][current]
            args.Y = Y[i][current]
            args.gamma = gamma[i]
            for k in rates:
                h_rates.append(k(args))   
                
        dt, index = smallest_event(h_rates)

        Xn, Yn, Zn = run_event(events[index%len(events)])
        
        mp = np.floor(index / len(events))
        
        for h in range(meta_pops):
            if(h == mp):
                X[h][current+1] = X[h][current] + Xn
                Y[h][current+1] = Y[h][current] + Yn
                Z[h][current+1] = Z[h][current] + Zn
            else:
                X[h][current+1] = X[h][current]
                Y[h][current+1] = Y[h][current]
                Z[h][current+1] = Z[h][current]
            N[h] = X[h][current+1] + Y[h][current+1] + Z[h][current+1]
    
        t += dt
        Ts[current+1] = t
        
        current += 1
        
    return Ts, X, Y, Z
```

```python
def create_rho(Rii, Rij, N):
    rho = np.zeros((N * N, N * N))
    for i in range(N * N):
        for j in range(N * N):
            row1 = i // N
            col1 = i % N
            row2 = j // N
            col2 = j % N
            if row1 == row2 and col1 == col2:
                rho[i][j] = Rii
            elif row1 == row2 and abs(col1 - col2) == 1:
                rho[i][j] = Rij
            elif col1 == col2 and abs(row1 - row2) == 1:
                rho[i][j] = Rij
    return rho
```

```python
beta = [1] * 2
gamma = [1/3] * 2

Rii = 1
Rij = 1/100

rho = create_rho(Rii, Rij, 2)
init = [(299, 1, 0), (300, 0, 0)]

t, x, y, z = gillespie_meta((m1, m2), events_meta, beta, gamma, rho, init, seed=4)

for i in range(2):
    plt.plot(t, y[i], label="Population " + str(i))
    
plt.title("Coupling of two metapopulations")
plt.plot([], [], ' ', label=r"$\beta$:" + str(beta[0]))
plt.plot([], [], ' ', label=r"$\gamma$:" + str(gamma[0])) 
plt.plot([], [], ' ', label=r"$\rho_{ii}$:" + str(Rii))
plt.plot([], [], ' ', label=r"$\rho_{ij}$:" + str(Rij)) 
plt.xlabel("Time (days)")
plt.ylabel("Number of infected people")
plt.legend()

print(np.where(y[1] != 0))
```

```python
def gillespie_lattice(rates, events, beta, gamma, rho, init, maxT=3000, seed=3):
    # set sed
    np.random.seed(seed)
    
    # Find smallest event
    def smallest_event(events):
        dts = np.array([])
        for event in events:
            if event == 0:
                dts = np.append(dts, math.inf)
            else:                
                dt = -1 / event * np.log(np.random.random_sample())
                dts = np.append(dts, dt)
        return(np.amin(dts), list(dts).index(np.amin(dts)))
    
    # execute event    
    def run_event(event):
        x_op, y_op, z_op = event
        return x_op, y_op, z_op
    
    # Total amount of populations    
    meta_pops = len(init)
    
    t = 0
    
    # Initialize arrays    
    Ts = np.full((maxT), None)    
    X = np.full((meta_pops, maxT), None)
    Y = np.full((meta_pops, maxT), None)
    Z = np.full((meta_pops, maxT), None)    
    
    N = np.full((meta_pops), None)
    
    terms = np.empty((meta_pops, 1))
    
    # Get last added element in arrays    
    current = 0
    
    # init    
    for i in range(meta_pops):
        X[i][0], Y[i][0], Z[i][0] = init[i]
        N[i] = sum(init[i])
        
    while t < maxT and current < maxT - 1:   
        # check wheter there are still infected people in metapops
        e = 0
        for i in range(meta_pops):
            if(Y[i][current] == 0):
                e += 1
        if(e == meta_pops):
            break

        h_rates = []            
        args = Args()
        
        for i in range(meta_pops):
            
            sumYi = 0
            sumYj = 0
            
            for j in range(meta_pops):
                sumYi += rho[j][i]
                sumYj += rho[i][j] * Y[j][current]
                
            args.term = (1 - sumYi) * Y[i][current] + sumYj

            args.X = X[i][current]
            args.Y = Y[i][current]
            args.N = N[i]
            args.beta = beta
            args.gamma = gamma
            
            for k in rates:
                h_rates.append(k(args))   
                
        dt, index = smallest_event(h_rates)

        Xn, Yn, Zn = run_event(events[index%len(events)])
        
        mp = np.floor(index / len(events))
        
        for h in range(meta_pops):
            if(h == mp):
                X[h][current+1] = X[h][current] + Xn
                Y[h][current+1] = Y[h][current] + Yn
                Z[h][current+1] = Z[h][current] + Zn
            else:
                X[h][current+1] = X[h][current]
                Y[h][current+1] = Y[h][current]
                Z[h][current+1] = Z[h][current]
            N[h] = X[h][current+1] + Y[h][current+1] + Z[h][current+1]
    
        t += dt
        Ts[current+1] = t
        
        current += 1
        
    return Ts, X, Y, Z
```
