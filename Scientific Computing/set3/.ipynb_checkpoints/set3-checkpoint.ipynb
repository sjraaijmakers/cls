```python
def to_1d(pos, n):
    x, y = pos
    return x*n + y

to_1d((1, 0), 3)
```

```python
def get_circle(n, L=1):
    mask = np.zeros((n, n))

    rad = L / 2
    
    for x in range(n):
        for y in range(n):
            if ((x - (n-1)/2) * (x - (n-1)/2) + (y - (n-1)/2) * (y - (n-1)/2) <= rad * rad): 
                mask[x][y] = 1

    M = np.zeros((n*n, n*n))
                
    for i in range(1, n):
        for j in range(1, n):
            current = i, j
            
            if mask[current] == 1:   

                left = i, j-1
                right = i, j+1
                top = i-1, j
                bottom = i+1, j
        
            
                if mask[right] == 1:
                    M[to_1d(current, n)][to_1d(right, n)] = 1
                    
                if mask[top] == 1:
                    M[to_1d(current, n)][to_1d(top, n)] = 1
                    
                if mask[left] == 1:
                    M[to_1d(current, n)][to_1d(left, n)] = 1
                    
                if mask[bottom] == 1:
                    M[to_1d(current, n)][to_1d(b, n)] = 1
                    
    print(M)
    
    return mask
        
            
            
            
                
                
```

```python
def get_m(n, m):
    M = np.zeros((n*m, n*m))
    
    for i in range(n*m):        
        t = 0
        
        # right
        if not ((i + 1) % m == 0):
            t += 1
            M[i][i + 1] = 1
        # left
        if not (i % m == 0):
            t += 1
            M[i][i - 1] = 1
        # top
        if not (i - m < 0):
            t += 1
            M[i][i - m] = 1
        # bottom
        if not (i + m >= n * m):
            t += 1
            M[i][i + m] = 1
            
        M[i][i] = -4
        
    return M
```

```python
def get_eig(n_M, L, N):
    N = L * N
    delta_x = L/N
    
    M = 1/delta_x**2 * n_M
    
    e_values, e_vectors = sp_linalg.eigsh(M)
    
    return np.real(e_values), e_vectors.T
```

```python
N = 50
square_m = get_m(N, N)
rect_m = get_m(N, 2*N)
```

```python
rect_m.shape
```

```python
square_e_values, square_e_vectors = get_eig(square_m, 1, N)
```

```python
for i in range(5):
    plt.title(np.sqrt(abs(square_e_values[i])))
    plt.imshow(np.abs(np.real(square_e_vectors[i].reshape((50, 50)))))
    plt.show()
```

```python
rect_e_values, rect_e_vectors = get_eig(rect_m, 1, N)
```

```python
for i in range(5):
    plt.title(np.sqrt(abs(rect_e_values[i])))
    plt.imshow(np.abs(np.real(rect_e_vectors[i].reshape((50, 100)))))
    plt.show()
```

```python
sq_e_values, sq_e_vectors = get_eig(square_m, 1, N)
```

```python
for i in range(5):
    plt.title(np.sqrt(-sq_e_values[i]))
    plt.imshow(np.abs(np.real(sq_e_vectors[i].reshape((N, N)))))
    plt.show()
```
