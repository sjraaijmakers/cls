```python
import matplotlib.pyplot as plt
import scipy as sp
from scipy import linalg
from tqdm import tqdm
import scipy.sparse.linalg as sp_linalg
import time
```

# 1

```python
def get_m(n, m):
    M = np.zeros((n*m, n*m))
    
    for i in range(n*m):        
        t = 0
        # right
        if not ((i + 1) % m == 0):
            t += 1
            M[i][i + 1] = 1
        # left
        if not (i % m == 0):
            t += 1
            M[i][i - 1] = 1
        # top
        if not (i - m < 0):
            t += 1
            M[i][i - m] = 1
        # bottom
        if not (i + m >= n * m):
            t += 1
            M[i][i + m] = 1
            
        M[i][i] = -4
        
    return M
```

```python
ex = get_m(3, 3)
print(ex)
```

```python
def get_eig(n_M, L, N, sort=False):
    delta_x = L/N
    
    M = 1/delta_x**2 * n_M
    
    e_values, e_vectors = sp.linalg.eigs(M)
    
    e_values = np.sqrt(np.abs(np.real(e_values)))
    e_vectors = e_vectors.T
    
    return e_values, e_vectors
```

```python
def sort_eigs(values, vectors):    
    l = sorted(zip(values, vectors), key = lambda x: x[0])
    return zip(*l)
```

### Square

```python
N = 50
square_m = get_m(N, N)
```

```python


square_e_values, square_e_vectors = get_eig(square_m, 1, N)
square_e_values, square_e_vectors = sort_eigs(square_e_values, square_e_vectors)
```

```python
fig, axes = plt.subplots(nrows=2, ncols=5, figsize=(13, 4), sharex='col', sharey='row')

for i in range(10):    
    ax = axes.flat[i]
    im = ax.imshow(square_e_vectors[i].reshape(50, 50), interpolation="spline36")
    ax.set_title(r"$\lambda$ = " + str(round(square_e_values[i], 2)))
    ax.grid('off')
    
fig.colorbar(im, ax=axes.ravel().tolist())
plt.savefig("square_eigenvectors.pdf")
```

### Circle

```python
def to_1d(pos, n):
    x, y = pos
    return x*n + y

def get_circle(N, L=1):
    mask = np.zeros((N, N))

    rad = L / 2
    
    mid = (N-1)/2
    
    for x in range(N):
        for y in range(N):
            if ((x - mid) * (x - mid) + (y - mid) * (y - mid) <= rad * rad): 
                mask[x][y] = 1

    M = np.zeros((N*N, N*N))
                
    for i in range(N):
        for j in range(N):
            current = i, j
            
            if mask[current] == 1:   

                left = i, j-1
                right = i, j+1
                top = i-1, j
                bottom = i+1, j
                
                neighbors = [right, top, left, bottom]
                
                candidates = []                
                for neighbor in neighbors:
                    if (neighbor[0] >= 0 and neighbor[0] < N) and (neighbor[1] >= 0 and neighbor[1] < N):
                        candidates.append(neighbor)
                    
                for c in candidates:
                    if mask[c] == 1:
                        M[to_1d(current, N)][to_1d(c, N)] = 1
                        
                M[to_1d(current, N)][to_1d(current, N)] = -4
                        
    return mask, M
```

```python
mask, circle_m = get_circle(50, 49)
```

```python
c_e_values, c_e_vectors = get_eig(circle_m, 1, N)
# remove for all eigenvalue of 0
c_e_values, c_e_vectors = sort_eigs(c_e_values[np.where(c_e_values > 0)], c_e_vectors[np.where(c_e_values > 0)])
```

```python
fig, axes = plt.subplots(nrows=2, ncols=5, figsize=(13, 4), sharex='col', sharey='row')

for i in range(10):    
    ax = axes.flat[i]
    im = ax.imshow(np.abs(c_e_vectors[i].reshape(50, 50)), interpolation="spline36")
    ax.set_title(r"$\lambda$ = " + str(round(c_e_values[i], 2)))
    ax.grid('off')
    
fig.colorbar(im, ax=axes.ravel().tolist())
plt.savefig("circle_eigenvectors.pdf")
```

```python
def get_eig_sh(n_M, L, N):
    delta_x = L/N
    
    M = 1/delta_x**2 * n_M
    
    e_values, e_vectors = sp_linalg.eigsh(sp.sparse.csr_matrix(M), k=10, which="SM")
    
    e_values = np.sqrt(np.abs(np.real(e_values)))
    e_vectors = e_vectors.T
    
    return e_values, e_vectors
```

```python
regular = []
sparse = []

for i in range(5):
    start = time.time()
    sh_val, _ = get_eig(square_m, 1, 50)
    regular.append(time.time() - start)

    start = time.time()
    sh_val, _ = get_eig_sh(square_m, 1, 50)
    sparse.append(time.time() - start)
```

```python
print("Avg time (rep = 5), for computing eigenvalues of square (50x50)")
print("Regular: " + str(np.mean(regular)))
print("Sparse: " + str(np.mean(sparse)))
```

## 1D

```python
def T(t, lam, A=1, B=1, c=1):
    return A * np.cos(c * lam * t) + B * np.sin (c * lam * t)
```

```python
from matplotlib import animation

def animate(images, name, lam, titles=None, interval=50, min2=0, max2=1):
    fig = plt.figure()

    container = []
    
    for i, img in enumerate(images):
        im = plt.imshow(img, animated=True, vmin=min2, vmax=max2, interpolation="spline36")
        plt.xlabel("x")
        plt.ylabel("y")
        title = plt.text(10.5,-3.05,r"$\lambda$ = " + str(lam) + ", T = " + str(titles[i]), horizontalalignment='center', verticalalignment='bottom')
        container.append([im, title])

    ani = animation.ArtistAnimation(fig, container, interval=interval, blit=False, repeat_delay=1000)

    ani.save(name)
```

```python
iteration = 0

ts = np.linspace(0, 2, 100)
times = T(ts, square_e_values[iteration])

first = square_e_vectors[iteration].reshape((50, 50))
vectors = []

for t in times:
    vectors.append(t * first)
    
titles = np.round(ts, 2)
animate(vectors, "square_50_l_" + str(round(square_e_values[iteration], 2)) + ".mp4", np.round(square_e_values[iteration], 2), titles=np.round(ts, 2), min2=np.amin(vectors), max2=np.amax(vectors))
```

```python
N = 100

m = get_m(N, N)
mask, _ = get_circle(N, N)

radius = 2
source = np.zeros((N, N))
midpoint = int(N/radius), int(N/radius)
source[midpoint[0] - int(N/2/radius * 1.2), midpoint[1] + int(N/2/radius * 0.6)] = 1

for i in range(N):
    for j in range(N):
        if source[i][j] == 1:
            m[i*N+j] = 0
            m[i*N+j][i*N+j] = 1
            
        if mask[i][j] == 0:
            m[i*N+j] = 0
            m[i*N+j][i*N+j] = 1
            
b = source.reshape((N*N))
c = sp_linalg.spsolve(m, b).reshape((N, N))

plt.imshow(c)
plt.xlabel("x")
plt.ylabel("y")
plt.colorbar()
```
