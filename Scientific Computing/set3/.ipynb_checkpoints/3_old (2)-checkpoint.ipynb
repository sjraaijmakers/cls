```python
import numpy as np
import scipy.linalg
import scipy.sparse.linalg
import matplotlib.pyplot as plt
import math
import timeit
```

```python
N = 99
m = N+2

# source
source = np.zeros((m,m))

mid = m/2
center = [m/2,m/2]
edge = (N)-m/2
dist_gridpoint = 2 / edge 


source[int(m/2 - np.round(1.2/dist_gridpoint, 0)), int(m/2 + np.round(0.6/dist_gridpoint, 0))] = 1

# circle
v = np.zeros((m,m))
## make a circle on the grid (v)
for i in range(N):
    for j in range(N):
        if math.sqrt(abs(center[0] - ((i+1)+0.5))**2 + abs(center[1] - ((j+1)+0.5))**2) > int(m/2)-1:
            v[i+1,j+1] = 1

# Make the matrix M: with Block structure and diagonal
M = np.zeros((N**2,N**2))
np.fill_diagonal(M, 4)

for i in range(N-1):
    M[i,i+1] = -1
    M[i+1,i] = -1

for i in range(N,N**2):
    if i%N != 0:
        M[(i-1),i] = -1
        M[i,(i-1)] = -1
    M[i,(i-N)] = -1
    M[(i-N),i] = -1

# get the target b vector, with current sums of knowns, and check for source and boundaries
b = np.zeros((N**2))

for i in range(N):
    for j in range(N):
        if source[i+1,j+1] == 1:
            M[i*N+j] = 0
            M[i*N+j][i*N+j] = 1
            b[i*N+j] = 1
            
        if v[i+1,j+1] == 1:
            M[i*N+j] = 0
            M[i*N+j][i*N+j] = 1
    
c = scipy.linalg.solve(M, b)

plt.imshow(c.reshape((99, 99)))
```
